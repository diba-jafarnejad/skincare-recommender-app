```python
# For textual columns, fill NaNs with empty strings
text_columns = ['product_name', 'brand_name', 'primary_category', 'skin_type', 'review_text', 'ingredients']
sephora_df[text_columns] = sephora_df[text_columns].fillna('')

# For numerical columns, fill NaNs with zeros
numerical_columns = ['rating', 'is_recommended', 'reviews', 'price_usd', 'total_pos_feedback_count']
sephora_df[numerical_columns] = sephora_df[numerical_columns].fillna(0)
```

```python
# clean Sephora cols with []
def clean_cell(x):
    if isinstance(x, float) and pd.isna(x):
        return ''
    if isinstance(x, str) and x.strip() == '[]':
        return ''
    if isinstance(x, list) and len(x) == 0:
        return ''
    return x

for col in text_columns:
    sephora_df[col] = sephora_df[col].apply(clean_cell)
```

```python
# aggregation
aggregation_functions = {
    'product_name': 'first',
    'brand_name': 'first',
    'ingredients': lambda x: list(set(ing for sublist in x.dropna() for ing in (sublist if isinstance(sublist, list) else [sublist]))),
    'rating': 'mean',
    'primary_category': 'first',
    'skin_type': lambda x: list(set(ty for sublist in x.dropna() for ty in (sublist if isinstance(sublist, list) else [sublist]))),
    'review_text': lambda x: list(x.dropna()),
    'is_recommended': 'mean',
    'reviews': 'first',
    'price_usd': 'mean',
    'total_pos_feedback_count': 'sum',
}

aggregated_df = sephora_df.groupby('product_id').agg(aggregation_functions).reset_index()
```
