# Create Ingredient-Based Features

```python
import ast

def safely_convert_to_list(val):
    try:
        if isinstance(val, str):
            parsed = ast.literal_eval(val)
            return parsed if isinstance(parsed, list) else []
        return val if isinstance(val, list) else []
    except:
        return []
```

```python
df['function_tags'] = df['function_tags'].apply(safely_convert_to_list)
df['who_should_avoid'] = df['who_should_avoid'].apply(safely_convert_to_list)
df['ingredients'] = df['ingredients'].apply(safely_convert_to_list)
```

```python
# Number of ingredients
df['num_ingredients'] = df['ingredients'].apply(lambda x: len(x) if isinstance(x, list) else 0)

# Number of functional tags
df['num_function_tags'] = df['function_tags'].apply(lambda x: len(set(x)) if isinstance(x, list) else 0)

# Flag: Has irritants (based on who_should_avoid)
df['has_irritants'] = df['who_should_avoid'].apply(lambda x: bool(x) if isinstance(x, list) else False)
```

#  Skin Match Score

```python
print(df.iloc[0]['function_tags'])
```

```python
import re

def parse_function_tags(raw_tags):
    if not isinstance(raw_tags, list):
        return []

    cleaned = []
    for tag_str in raw_tags:
        if isinstance(tag_str, str):
            # Remove square brackets, split by comma, strip whitespace
            tag_list = re.sub(r'[\[\]]', '', tag_str).split(',')
            for tag in tag_list:
                tag = tag.strip().lower()
                if tag and tag != 'none':
                    cleaned.append(tag)
    return list(set(cleaned))  # remove duplicates
```

```python
df['parsed_function_tags'] = df['function_tags'].apply(parse_function_tags)
```

```python
# Broader and more specific skin type matching
skin_type_tag_map = {
    "dry": ["hydrating", "barrier-repair", "healing", "moisturizing"],
    "oily": ["oil-control", "exfoliating", "acne-fighting"],
    "sensitive": ["soothing", "anti-redness", "anti-inflammatory"],
    "acne-prone": ["acne-fighting", "oil-control", "anti-inflammatory"],
    "mature": ["anti-aging", "firming", "antioxidant"],
    "dull": ["brightening", "antioxidant", "exfoliating"]
}

def skin_match_score(flat_tags, user_skin):
    return sum(1 for tag in flat_tags if user_skin in skin_type_tag_map.get(tag, []))
```

```python
user_skin_type = 'dry'  # or from user input

df['skin_match_score'] = df['parsed_function_tags'].apply(lambda x: skin_match_score(x, user_skin_type))
```

```python
df[df['skin_match_score']!=0]
```

```python
def get_matching_tags(tags, user_skin):
    return [tag for tag in tags if user_skin in skin_type_tag_map.get(tag, [])]

df['matched_tags'] = df['parsed_function_tags'].apply(lambda x: get_matching_tags(x, user_skin_type))
```

# Popularity-Based Features

```python
df.columns
```

```python
from sklearn.preprocessing import MinMaxScaler

# Define columns to scale
to_normalize = [
    'rating',
    'is_recommended',
    'reviews',
    'price_usd',
    'total_pos_feedback_count',
    'skin_match_score',
    'num_ingredients',
    'num_function_tags'
]

# Fill NaNs just in case
df[to_normalize] = df[to_normalize].fillna(0)

# Initialize and apply scaler
scaler = MinMaxScaler()
df_scaled = pd.DataFrame(scaler.fit_transform(df[to_normalize]),
                         columns=[f"{col}_scaled" for col in to_normalize])

# Add to original df
df = pd.concat([df, df_scaled], axis=1)
```

```python
df
```

# Create Final Relevance Score

```python
df['relevance_score'] = (
    df['skin_match_score_scaled'] * 0.4 +
    df['rating_scaled'] * 0.25 +
    df['is_recommended_scaled'] * 0.15 +
    df['total_pos_feedback_count_scaled'] * 0.1 +
    df['num_function_tags_scaled'] * 0.1
)
```
